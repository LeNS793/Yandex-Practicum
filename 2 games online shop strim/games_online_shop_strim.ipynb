**Комментарии**  

Возможные причины пропусков:
- _name_ (категориальная переменная, тип пропусков - полностью случайные): данный параметр очень важен - без него вообще не имеет смысла вносить игру в датасет. Он пропущен явно случайно, восстановить возможности нет, кол-во строк небольшое - поэтому удаляем;
- _genre_ (категориальная переменная, тип пропусков - случайные): пропуски могут зависеть от столбца `name` - возможно игра какая-то специфическая, для которой нет подходящего жанра. Можно попытаться восстановить, если разобрать отдельно каждый случай, но это займент много времени. И т.к. кол-во пропусков тут небольшое, то можно эти строки спокойно удалить;
- _year_of_release_ (количественная переменная, тип пропусков - полностью случайные): если у нас есть название игры, то эти данные можно было бы указать. Т.е. пропуски в этом столбце не зависят ни от каких других значений. Пропуски в этом столбце составляют 1,6 % от общего набора датасета, поэтому их удаляем;  
- _rating_ (категориальная переменная, тип пропусков - неслучайные: их наличие зависит от значений в других столбцах, например рынок или год выпуска): организация ESRB была основана в 1994 году для присвоения возрастного рейтинга игр, а свое окончательное название она приобрела в 2003 году. Поэтому все пропуски в данном столбце у игр, вышедших до 2003 года - это норма. Этим можно объяснить примерно 40 % всех пропусков. Остальные 60% - рейтинг данной организации действителен только для продаж в США и Канаде. Т.е. если для игры будет целевой рынок - Европа, то данный рейтинг ей не требуется;
- _critic_score_ (количественная переменная, тип пропусков - случайные, можно объяснить с помощью данных из других столбцов): приблизительно до 1995 года выходило не так много игр - элементарно нечего было оценивать, отрасль была не развита. Например, в 1988 году из 15 игр оценку от критиков 64 балла получила только одна игра, которая продалась по миру тиражом всего 30 000 экземпляров. После 1995 года ситуация изменилась: отрасль более-менее развилась, игры начали выходить в достаточно большом кол-ве. Тут сыграли законы рынка: те игры, у которых были достаточно большие продажи, были хорошо оценены критиками - они вообще обратили на них внимание (хотя в последнее время иногда бывает, что критикам позволяют тестировать игры до релиза). А для игр с низкими продажами - отсутствие оценок = пропуски. Если мы попытаемся заполнить пропуски средним значением или медианой, то можем создать аномалию, что приведет к искажению результатов исследования. По этим причинам мы не можем заполнить имеющиеся пропуски в данном столбце - оставим их без изменений;
- _user_score_ (количественная переменная, тип пропусков - случайные): ситуация, аналогичная данным из столбца `critic_score`. До 1995 года игр было мало (менее 200 штук), после 1995 плохие игры просто не попадали в поле зрения игроков. Связь тут очевидна - игры с малым кол-вом продаж не оцениваются. Пропуски не удаляем. Что касается значений `'tbd'`, то в игровой индустрии это может означать To Be Determined, то есть "Будет определено". Очень часто подобное сокращение можно встретить в анонсах компьютерных игр в графе "Дата выхода". Когда разработчик той или иной игры уже сообщил о том, что она готовится к выходу, но пока еще не может назвать точную дату, именно это сокращение и используется. Поэтому замена данных значений на NaN допускается - мы не потеряем никакие данные. 

Полных дубликатов строк не обнаружено.

### Исследование влияния оценок пользователей и критиков на объемы продаж игр

```python
# исключим все игры жанра sports

(
    afd
    .query('platform == "PS4" and genre != "Sports"')
    .plot(kind='scatter', y='total_sales' , x='user_score', alpha=0.5, subplots=True, figsize=(10,6), c = 'b', s = 4)
)   

plt.title('Диаграмма рассеяния — Оценка игроков — объем продаж');

pearson_correlation1 = round(afd[afd['platform'] == 'PS4']['user_score'].\
                       corr(afd[afd['platform'] == 'PS4']['total_sales']), 2)
    
print(f'Коэффициент Пирсона, показывающий взаимосвязь изменения объема продаж игр от оценки игроков, \
составляет {pearson_correlation1}\n')
```

**Выводы:** 
Диаграмма рассеяния - Оценки игроков - объем продаж: тут парадоксальная ситуация - есть игры с высоким рейтингом от критиков и достаточно большим уровнем продаж, однако оценки игроков относительно низкие. В основном это симуляторы игровых видов спорта (футбол и хоккей). Причем объяснить этот феномен затруднительно, т.к. критикам нет смысла завышать оценку - данные виды спорта итак достаточно разрекламированы. И именно из-за таких игр корреляция "оценка игроков - объем продаж" отсутствует.

### Самые популярные платформы по регионам

```python
# снова использовали фильтрацию по годам, т.к. до этого были отброшены неактуальные на первый
# взгляд платформы. Также у нас были платформы из группы риска, по которым нужно было принять
# решение - закупаем ли мы игры на этот год

afd_yor = afd[afd['year_of_release'] > 2012] #.query('year_of_release > 2012') 

# получен объект типа Index со списком регионов

countries = afd_yor.columns[4: 7]

# создаем несколько графиков с помощью функции subplots, в fig записываем кол-во областей, 
# где будем создавать графики (1 область), в axs записываем кол-во графиков (3 "пирога")

fig, axs = plt.subplots(1, len(countries), figsize = (15, 5))

fig.suptitle('Топ-5 платформ в регионах с 2013 года')

# строим "пирог" для каджого региона, перебирая регионы через цикл

for i in range(len(countries)):
    
    # группируем датафрейм по платформам, выбираем регион и суммируем значения
    # (по сути в этом столбце находится информация о продажах игр)
    
    circle = (afd_yor.groupby('platform')[countries[i]].sum()
           .reset_index()
           .sort_values(countries[i], ascending = False, ignore_index = True)        
           .head(5)) # записываем в переменную 7 популярных игровых платформ
   
    # добавляем к переменной еще одну строку с названием платформы'other_platforms'
    # в регион записываем разницу между общей суммой продаж и суммой продаж на 7 платформ 
    
    # строку добавляем с помощью логической индексации. Метод append не используем, т.к. в будущем он
    # прекратит функционировать
    
#     circle = circle.append({'platform': 'other_platforms',\
#                             countries[i]: afd_yor[countries[i]].sum() - circle[countries[i]].sum()},\
#                             ignore_index=True)
    
    circle.loc[len(circle.index)] = ['other_platforms', afd_yor[countries[i]].sum() - circle[countries[i]].sum()]
    
    # строим круговую диаграмму под номером i для данных из circle, добавляем текстовые метки
    # c помощью labels и выводим процент с помощью autopct
    
    axs[i].pie(circle[countries[i]], labels = circle['platform'], autopct = '%1.1f%%')
    axs[i].set_title(circle.columns[1].split('_')[0].upper())
plt.show()
```

**Вывод:** Также нужно учитывать, что эти данные показывают ситуацию за прошлое время, но не учитывают тренды, направления движения рынка сейчас и сроки жизни платформ. ## Проверка гипотез

### Сравнение средних пользовательских рейтингов платформ Xbox One и PC

Гипотезы:
 - нулевая Н0: средние пользовательские рейтинги платформ Xbox One и PC `равны`;
 - альтернативная Н1: средние пользовательские рейтинги платформ Xbox One и PC `не равны`.

Проверим гипотезы: 

```python
sample_xone = afd.query('platform == "XOne" and user_score.notnull()')['user_score']
sample_pc = afd.query('platform == "PC" and user_score.notnull()')['user_score']

results = st.ttest_ind(
     sample_xone,
     sample_pc)

alpha = 0.05

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
    
print()    
print('Среднее значение пользовательского рейтинга для игр платформы Xbox One равно',\
     round(afd[afd['platform'] == 'XOne']['user_score'].mean(),1))
print('Среднее значение пользовательского рейтинга для игр платформы PC равно',\
     round(afd[afd['platform'] == 'PC']['user_score'].mean(),1))
```

**Комментарии:** Здесь мы сравниваем средние значения генеральных совокупностей (которые получаем церез ЦПТ из нашей фильтрованной таблицы), поэтому используем метод scipy.stats.ttest_ind(). Пропуски из данных столбцов исключаем в момент фильтрации, чтобы метод корректно рассчитал p-value. 

### Сравнение средних пользовательских рейтингов жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования»)

Гипотезы:
 - нулевая Н0: средние пользовательские рейтинги жанра Action и Sports `равны`;
 - альтернативная Н1: средние пользовательские рейтинги жанра Action и Sports `не равны`.

Проверим гипотезы: 

```python
sample_action = afd.query('genre == "Action" and user_score.notnull()')['user_score']
sample_sports = afd.query('genre == "Sports" and user_score.notnull()')['user_score']

results = st.ttest_ind(
     sample_action,
     sample_sports)

alpha = 0.05

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")

print()    
print('Среднее значение пользовательского рейтинга для игр жанра Action равно',\
     round(afd[afd['genre'] == 'Action']['user_score'].mean(), 1))
print('Среднее значение пользовательского рейтинга для игр жанра Sports равно',\
     round(afd[afd['genre'] == 'Sports']['user_score'].mean(), 1))
```

**Комментарии:** На имеющихся данных, на уровне значимости 5% есть основания отвергнуть нулевую гипотезу о равенстве средних значений пользовательских рейтингов (отзывов) жанров Action и Sports в пользу альтернативной гипотезы.
