#### Поиск пропущенных значений

```python
initial_data.isna().sum()
```

```python
# визуальное представление пропущенных значений

def pass_value_barh(df):
    try:
        (
            (df.isna().mean()*100) # получаем % кол-ва пропусков от общего значения
            .to_frame() # преобразуем объект Series в DataFrame (он получается с одним столбцом)
            .rename(columns = {0:'space'}) # в полученном DataFrame переименовываем столбец, в качестве индексов остаются
                                           # названия столбцов. Видимо 0 - имя нового столбца по умолчанию
            .query('space > 0')            # выбираем все столбцы, где присутствуют пропуски
            .sort_values(by = 'space', ascending = True) # сортируем таблицу по убыванию кол-ва пропусков 
            .plot(kind = 'barh', figsize = (19,6), rot = -5, legend = False, fontsize = 16) # строим горизонтальную
                                                                                            # столбчатую диаграмму 
            .set_title('Пропущенные значения в % от общего объема данных' + "\n", fontsize = 22, color = 'SteelBlue') # задаем заголовок для графика    
        );    
    except:
        print('пропусков не осталось :) или произошла ошибка в первой части функции ')
```

```python
pass_value_barh(initial_data)
```

#### **Анализируем данные из столбца 'first_day_exposition' - дата публикации**

```python
# сделаем перезапись исходных данных в новую переменную. Возможно к исходным данным еще придется обращаться

filtered_data = initial_data 

filtered_data['first_day_exposition'] = pd.to_datetime(filtered_data['first_day_exposition'],\
                                                         format = '%Y-%m-%dT%H:%M:%S') 
filtered_data['first_day_exposition'].unique()
```

**Комментарий:** меняем тип данных для последующего корректного использования, создания срезов по дням и месяцам.

#### **Анализируем данные из столбца 'ceiling_height' - высота потолков**

```python
print(filtered_data['ceiling_height'].unique())
```

```python
ceiling_height_median = initial_data.query('ceiling_height <= 4')['ceiling_height'].median()
filtered_data['ceiling_height'] = filtered_data['ceiling_height'].fillna(ceiling_height_median)

filtered_data.loc[filtered_data['ceiling_height'] > 14, 'ceiling_height'] =\
    filtered_data.loc[filtered_data['ceiling_height'] > 14, 'ceiling_height'] / 10 

filtered_data['ceiling_height'].unique()
```

**Комментарий:** Пропусков много, но данный параметр не очень критичен, поэтому пропущенные значения были заменены на медианное - 2,65 м, которое было рассчитано по первоначальным данным. Замена значений сделана с помощью цикла и словаря, а после удалены нереалистичные значения с помощью среза.

#### **Анализируем данные из столбца 'floors_total' - всего этажей в доме** 

```python
filtered_data = filtered_data.dropna(subset = ['floors_total'])
filtered_data = filtered_data.reset_index(drop = True)
filtered_data['floors_total'] = filtered_data['floors_total'].astype('int')
```

**Комментарий:** кол-во строк с пропусками не такое большое (86), поэтому их можно удалить - на результаты исследования это не повлияет. Есть достаточно много значений больше 30, но они в основном в Санкт-Петербурге, где присутствует многоэтажная застройка. Поэтому данные строки не удалены. 

#### **Анализируем данные из столбца 'living_area' и 'kitchen_area' - жилая площадь и площадь кухни**

```python
# есть строки, в которых жилая площадь больше, чем общая площадь (или 0.9 от общей площади). Это явная ошибка 
# данных и следует ее исправить до того, как заполнять пропуски в площадях, иначе у нас получатся отрицательные
# значения в заполняемой площади. Ищем соотношение жилой и кухни к общей, причем ищем по тем строкам, в которых соотношение
# площадей корректное:

percentage_table = filtered_data.query('total_area * 0.9 > (living_area + kitchen_area) & \
                                        total_area * 0.9 > living_area &\
                                        total_area * 0.9 > kitchen_area')

# в таблице percentage_table соотношение площадей жилая/кухня/общая правильное, соответствует фильтру. Из строк этой
# таблицы и будем считать % соотношения через медиану

percent1 = round(percentage_table['living_area'].median() / percentage_table['total_area'].median(), 2)
percent2 = round(percentage_table['kitchen_area'].median() / percentage_table['total_area'].median(), 2)
```

```python
# исправляем неверные соотношения площадей, используя полученные медианы соотношений жилой/общей и кухни/общей площадям
# для total_area * 0.9 < (living_area + kitchen_area)

filtered_data.loc[filtered_data['total_area'] * 0.9 < filtered_data['living_area'] + filtered_data['kitchen_area'],\
                  'living_area'] = \
round(filtered_data.loc[filtered_data['total_area'] * 0.9 < filtered_data['living_area'] + filtered_data['kitchen_area'],\
                        'total_area'] * percent1, 1)     
        
filtered_data.loc[filtered_data['total_area'] * 0.9 < filtered_data['living_area'] + filtered_data['kitchen_area'],\
                  'kitchen_area'] = \
round(filtered_data.loc[filtered_data['total_area'] * 0.9 < filtered_data['living_area'] + filtered_data['kitchen_area'],\
                        'total_area'] * percent2, 1)

# теперь для total_area * 0.73 < living_area (иначе в kitchen_area будут отрицательные значения)


filtered_data.loc[filtered_data['total_area'] * 0.73 < filtered_data['living_area'] ,'living_area'] = \
round(filtered_data.loc[filtered_data['total_area'] * 0.73 < filtered_data['living_area'], 'total_area'] * 0.727, 1) 

        
# заполняем жилую площадь для квартир при условии, что и жилая, и площадь кухни пропущены - через процентное соотношение

filtered_data.loc[(filtered_data['living_area'].isna()) & (filtered_data['kitchen_area'].isna()), 'living_area'] = \
round(filtered_data.loc[(filtered_data['living_area'].isna()) & (filtered_data['kitchen_area'].isna()),'total_area']\
      * percent1, 1)
 
# заполняем жилую площадь при условии, что пропущена только жилая площадь, а площадь кухни заполнена
# через соотношение living_area = total_area * 0.89 - kitchen_area. Берем 0.89, т.к. округление через round() происходит в  
# большую сторону и у нас появляются квартиры, которые выходят за рамки соотношений площадей
# использовать int() и math.floor() для округления в меньшую сторону не получилось - выдает ошибку

filtered_data.loc[filtered_data['living_area'].isna(), 'living_area'] = \
round(filtered_data.loc[filtered_data['living_area'].isna(),'total_area'] * 0.89 -\
      filtered_data.loc[filtered_data['living_area'].isna(),'kitchen_area'], 1)

# заполняем площадь кухни для студий - везде, где studio == True, kitchen_area - NaN, поэтому заполняем нулями
# после выполнения кода к исходным 473 объявлениям, которые были изначально, прибавились еще 3, 
# в которых total_area * 0.9 < living_area, но они не могли быть учтены из-за значений NaN   

filtered_data.loc[(filtered_data['kitchen_area'].isna()) & (filtered_data['studio'] == True), 'kitchen_area'] = 0

# заполняем площадь кухни через соотношение kitchen_area = total_area * 0.89 - living_area 

filtered_data.loc[filtered_data['kitchen_area'].isna(), 'kitchen_area'] = \
round(filtered_data.loc[filtered_data['kitchen_area'].isna(),'total_area'] * 0.89 -\
      filtered_data.loc[filtered_data['kitchen_area'].isna(),'living_area'], 1)
```

**Комментарий:** т.к. для всех объектов у нас известна общая площадь, то мы можем заполнить пропуски для жилой площади и кухни. Но нужно учитывать, что у нас могут быть следующие ситуации:
- не заполнены жилая и кухня;
- заполнена жилая, не заполнена кухня;
- не заполнена жилая, заполнена кухня.

Проблемы с соотношениями площадей:
- общая * 0.9 < жилая + кухня;
- общая * 0.73 < жилая;
- общая * 0.17 < кухня.

Все эти проверки мы провели и исправили перед заполнением пропусков, чтобы при заполнении кухонной площади мы не получали отрицательных или нулевых значений (для квартир). За самые надежные и важные данные принимаем общую площадь (от нее напряму зависит стоимость квартиры и ее стараются вносить правильно), затем идет жилая, а относительно не важная - площадь кухни. 

С помощью медианных значений узнали, сколько в % от общей занимают жилая и площадь кухни - 58 % и 17 % соответственно.  Данные реалистичны (в идеале было бы заполнить медианами по кол-ву комнат, т.к. соотношение меняется).

Коэффициент площади для комнат 0.73 взят вместо 0.58 как разница (0.9 - 0.17) - соотношение кухни|общей не так критично, а соотношение жилой/общей более важно и поэтому не ограничиваем его медианным 0.58. А т.к. у нас почти половина датасета не подходит под условие `общая * 0.17 < кухня`, то стараемся площадь кухни не корректировать.

Сначала заполнили жилую и кухонную площади в тех строках, где оба значение пропущены. Затем привели жилую площадь к соотношению `общая * 0.73 > жилая`. И остались пропуски для `жилой площади`в третьем варианте, когда пропущена только она, а общая и кухонная площади заполнены.   

Для студий везде площадь кухни пропущена - заполняем нулями.

Для заполнения была применена логическая индексация, т.к. она быстрее функций.

Также здесь за выбросы можно принять квартиры с большой площадью. Но в нашем случае они не являются выбросами, а скорее объявлениями о продаже премиальной недвижимости (была построена диаграмма размаха и просмотрены данные по квартирам большой площади). По хорошему тут следовало бы ввести сегменты деления недвижимости, например "эконом", "стандрат", "премиум".

```python
def separate_on_floors(row):
    floor = row['floor']
    floors_total = row['floors_total']
    
    if floor == 1:
        return 'первый'
    elif floor == floors_total:
        return 'последний'
    else:
        return 'другой'

filtered_data2['floor_type'] = filtered_data2.apply(separate_on_floors, axis = 1)

type = 'последний'
```

Было принято решение проводить исследовательский анализ в разрезе кол-ва комнат, т.к. этот параметр может очень сильно влиять на остальные показатели. Отдельно были проанализированы квартиры-студии. Как в дальнейшем показали данные, поведения квартир с разным кол-вом комнат оказались практически идентичны, кроме величин самих значений (например общая стоимость).

Основные тезисы:

- типичный портрет квартиры: общая площадь 10-115 м2, жилая - не более 80  м2, кол-во комнат - не больше 7, этажность застройки - 28 этажей;
- на рынке было больше всего объявлений о продаже одно- и двухкомнатных квартир (30 % и 28 % соответственно). Также достаточно большую долю составляют трехкомнатные квартиры (22-23 %);
- кол-во квартир на первом и последнем этажах составляют 11 % и 12 % от выборки;
- прослеживается четкая сезонность в продажах: пиковые периоды размещения объявлений 
- цены на недвижимость в период 2014 - 2019 гг. по разным классам недвижимости практически не изменились, а в середине этого периода даже снижались;
- студии прекратили продавать с 2017 года - в датасете нет ни одного объявления;
- продажу квартиры можно считать быстрой, если она произошла в течение 45 дней, нормальной - до 3 месяцев, `долгой - 430 дней`;
- самая высокая стоимость 1 м2 в период 2014-2019 гг. была в Санкт-Петербурге и составила 116835.84 рублей (для однокомнатной квартиры);

- по некоторым категориям квартир не достаточно кол-ва объявлений, чтобы сделать объективные выводы (например однокомнатные квартиры на расстоянии до 3 км до центра Санкт-Петербурга);
- с 2014 по 2019 гг. срок размещения объявлений сильно снизился: с 600 дней в 2014 до 30 дней в 2019.
